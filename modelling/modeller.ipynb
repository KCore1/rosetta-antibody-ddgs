# Modelling whole protein
NOTE: I did remove a few residues at the end of the strain so that the alignment would not try to match with residues on the antibody and remodel those. This caused the alignment to cover purely the hemagglutinin.

```python
# Homology modelling by the automodel class for CR6261-H1
from modeller.automodel import * # Load the automodel class
log.verbose() # request verbose output
env = Environ() # create a new MODELLER environment
env.io.atom_files_directory = ['/projects/kach6913/rosetta-antibody-ddgs/modelling/']
a = AutoModel(env,
    alnfile = '/projects/kach6913/rosetta-antibody-ddgs/modelling/ncal-3GBN.ali', # alignment filename
    knowns = '3GBN_all', # codes of the templates
    sequence = 'New_Caledonia_1999') # code of the target
a.starting_model= 1 # index of the first model
a.ending_model = 1 # index of the last model
# (determines how many models to calculate)
a.make() # do the actual homology modelling
```

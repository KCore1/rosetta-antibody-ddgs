# MODELLER
Using modeller to homology model the flu antibody pdbs, as there are a few mutational differences in the stem that we may not capture.

```python
from modeller import *
from modeller.automodel import *
```

```python
cd /projects/kach6913/rosetta-antibody-ddgs/modelling/
```

```python
# This will read two sequences, align them, and write the alignment
# to a file:
log.verbose()
env = Environ()
env.io.atom_files_directory = ['/projects/kach6913/rosetta-antibody-ddgs/modelling/']
aln = alignment(env)
mdl = model(env, file='3GBN_all')
aln.append_model(mdl, align_codes='3GBN_all')
aln.append(file='/projects/kach6913/rosetta-antibody-ddgs/modelling/H1_PDB_toalign.seq', align_codes=('New_Caledonia_1999'))
# The as1.sim.mat similarity matrix is used by default:
aln.align(gap_penalties_1d=(-600, -400))
aln.write(file='/projects/kach6913/rosetta-antibody-ddgs/modelling/ncal-3GBN.ali', alignment_format='PIR')
aln.write(file='/projects/kach6913/rosetta-antibody-ddgs/modelling/ncal-3GBN.pap', alignment_format='PAP')
```

```python
# TESTING if selection works
s = Selection()
s.add(mdl.chains['A'])
print(s)
```

```python
class MyModel(AutoModel):
    def select_atoms(self):
        return Selection(self.residue_range('1:A', '2:A'))
```

```python
# Homology modelling by the automodel class
from modeller.automodel import * # Load the automodel class
log.verbose() # request verbose output
env = Environ() # create a new MODELLER environment
env.io.atom_files_directory = ['/projects/kach6913/rosetta-antibody-ddgs/modelling/']
a = MyModel(env,
    alnfile = '/projects/kach6913/rosetta-antibody-ddgs/modelling/ncal-3GBN.ali', # alignment filename
    knowns = '3GBN_all', # codes of the templates
    sequence = 'New_Caledonia_1999') # code of the target
a.starting_model= 1 # index of the first model
a.ending_model = 1 # index of the last model
# (determines how many models to calculate)
a.make() # do the actual homology modelling
```
